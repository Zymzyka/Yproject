```python
#Уникальные значения рейтинагов ESBR
df['rating'].unique()
```

```python
#Делаем значения удобного вида, для восприятия
def rating_name (row):
    try:
        if 'E10' in row:
            return 'от 10 лет'
        elif 'E' in row or 'K-A' in row:
            return 'Любой возраст'
        elif 'E10+' in row:
            return 'от 10 лет'
        elif 'T' in row:
            return 'От 12 лет'
        elif 'M' in row :
            return 'От 17 лет'
        elif 'EC' in row:
            return 'Раннее детство'
        elif 'RP' in row:
            return 'Ожидается рейтинг, примерно 17+'
        elif 'AO' in row:
            return 'Только для взрослых'
    except:
        return 'нет рейтинга'
           
```

```python
#Сделали новую колонку со значениями отсортированными
df['rating_name'] = df['rating'].apply(rating_name)
df['rating_name'].unique()
```

<div class="alert alert-block alert-info">
Сделали столбец rating_name с более простым понимаем классификации рейтинга, также были обнаружены дубликаты это рейтинг E и K-A, K-А cо временем упрозднили и теперь все игры маркируются рейтингом E , было принято решение эти рейтинги в таблице обьединить.

H0: средний пользовательский рейтинг платформ Xbox и PC одинаковый

H1: средний пользовательский рейтинг жанров Xbox и PC различается


```python
# Cоздаем две генеральные совокупности по условию задачи
xone = df_curr[(df_curr['platform'] == 'xone') & (df_curr['user_score'] > 0)]['user_score']
pc = df_curr[(df_curr['platform'] == 'pc') & (df_curr['user_score'] > 0)]['user_score']
```

```python
alpha = 0.05
results = st.ttest_ind(xone, pc, equal_var = False)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

# Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

H0: средний пользовательский рейтинг жанров Action и Sports разные

H1: средний пользовательский рейтинг жанров Action и Sports одинаковые


```python
# Cоздаем две генеральные совокупности по условию задачи
action = df_curr[(df_curr['genre'] == 'action') & (df_curr['user_score'] > 0)]['user_score']
sports = df_curr[(df_curr['genre'] == 'sports') & (df_curr['user_score'] > 0)]['user_score']
```

```python
alpha = 0.05
results = st.ttest_ind(action, sports, equal_var = False)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
